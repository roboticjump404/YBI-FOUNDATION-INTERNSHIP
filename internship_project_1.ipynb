MILEAGE PREDICTION

OBJECTIVE :
To predict the mileage of vehicles


Data Source: "https://github.com/YBI-Foundation/Dataset/raw/main/MPG.csv"

```python
#import library
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#import data
Mil=pd.read_csv("https://github.com/YBI-Foundation/Dataset/raw/main/MPG.csv")
Mil.head()
```

```python
Mil.columns
```

```python
#remove missing values
Mil=Mil.dropna()
```

```python
#describe data
Mil.info()
```

```python
Mil.describe()
```

```python
#data visualisation
sns.regplot(x='displacement',y='mpg',data=Mil)
```

```python
#data preprocessing
Mil.info()
```

```python
Mil.corr()
```

```python
Mil.describe()
```

```python
#define target variable y and feature x
Mil.columns
```

```python
y=Mil['mpg']
x=Mil[['displacement', 'horsepower', 'weight',
       'acceleration']]
```

```python
#train test split
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y, train_size=0.8, random_state=2)
```

```python
x_train.shape, x_test.shape, y_train.shape, y_test.shape
```

```python
#linear regression model
from sklearn.linear_model import LinearRegression
model=LinearRegression()
```

```python
#modeling
model.fit(x_train,y_train)
```

```python
model.coef_
```

```python
model.intercept_
```

```python
#Predict model
y_pred=model.predict(x_test)
```

```python
y_pred
```

```python
# model accuracy and evaluation
from sklearn.metrics import mean_absolute_percentage_error
mean_absolute_percentage_error(y_test,y_pred)
#hence model is 87% accurate
```
